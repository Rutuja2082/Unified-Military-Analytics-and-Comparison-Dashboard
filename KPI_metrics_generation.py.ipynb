```python
df = pd.read_excel("Military_Cleaned_data.xlsx")
df.to_csv("Military_Cleaned_data.csv", index=False, encoding="utf-8")
```

```python
df.info()
```

```python
# 1 Assets per capital (total_aircraft + tanks + naval_assets) / population

df["Assets_Per_Capita"] = (
    df["Aircraft_Total"]
    + df["Armor_Tanks_Total"]
    + df["Navy_Ships"]
) / df["Total_Population_By_Country"]
df["Assets_Per_Capita"].head()
```

```python
# KPI 2: Defense Budget to GDP Ratio
# Defense Budget- Defense_Spending_Budget
# GDP- Purchasing_Power_Parity
#GlobalFirepower does not provide direct GDP
#Purchasing_Power_Parity (PPP) is used as a GDP proxy, which is standard in defense analytics

df["Defense_Budget_to_GDP_Ratio"] = (
    df["Defense_Spending_Budget"] /
    df["Purchasing_Power_Parity"]
)
df["Defense_Budget_to_GDP_Ratio"].head()
```

```python
#Active_Military_Manpower 
# Active_Reserve_Military_Manpower 
df["Total_Personnel"] = df["Active_Military_Manpower"] + df["Active_Reserve_Military_Manpower"]
df["Total_Personnel"].head(10)
```

```python
# KPI 3: Personnel Density

df["Personnel_Density"] = (
    df["Total_Personnel"] /
    df["Total_Population_By_Country"]
)
df["Personnel_Density"].head(10)
```

```python
# KPI 4: Defense Budget per Soldier

df["Defense_Budget_Per_Soldier"] = (
    df["Defense_Spending_Budget"] /
    df["Active_Military_Manpower"]
)
df["Defense_Budget_Per_Soldier"].head(10)
```

```python
# KPI 5: Power Index Rank Gap (Reference: USA Rank = 1)

REFERENCE_RANK = 1

df["Power_Index_Rank_Gap"] = (
    df["Power_Index_Rank"] - REFERENCE_RANK
)
df["Power_Index_Rank_Gap"].head(10)
```

```python
# KPI 6: Air Power Ratio

df["Air_Power_Ratio"] = (
    df["Aircraft_Total"] / df["Active_Military_Manpower"]
)
df["Air_Power_Ratio"].head(10)
```

```python
# KPI 7: Armor Intensity Index

df["Armor_Intensity_Index"] = (
    df["Armor_Tanks_Total"] / df["Square_Land_Area"]
)
df["Armor_Intensity_Index"].head(10)
```

```python
df = df.rename(columns={
    "Navy_Ships": "Naval_Assets"
})
```

```python
# KPI 8: Naval Strength per Coastline

df["Naval_Strength_Per_Coastline"] = (
    df["Naval_Assets"] / df["Coastline_Coverage"]
)
df["Naval_Strength_Per_Coastline"].head()
```

```python
# KPI-9: Military Burden Index
df["Military_Burden_Index"] = (
    df["Defense_Spending_Budget"] / df["Total_Population_By_Country"]
)
df["Military_Burden_Index"].head() 
```

```python
#Create Total Assets per Country

df["Total_Military_Assets"] = (
    df["Aircraft_Total"]
    + df["Armor_Tanks_Total"]
    + df["Naval_Assets"]
)
df["Total_Military_Assets"].head()
```

```python
# 10 Coalition Strength (Example: Selected Countries)
selected_countries = ["United States", "United Kingdom", "France", "Germany"]

coalition_strength = df[
    df["Country"].isin(selected_countries)
]["Total_Military_Assets"].sum()

coalition_strength
```
